```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
x, y = train, np.ravel(trainLabel)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer

std = StandardScaler()
x_std = std.fit_transform(x)
mms = MinMaxScaler()
x_mms = mms.fit_transform(x)
norm = Normalizer()
x_norm = norm.fit_transform(x)
```

```python
from sklearn.mixture import GaussianMixture

x_all = np.r_[train,test]
lowest_bic = np.infty
bic = []
n_components_range = range(1, 7)
cv_types = ['spherical', 'tied', 'diag', 'full']
for cv_type in cv_types:
    for n_components in n_components_range:
        gmm = GaussianMixture(n_components=n_components,covariance_type=cv_type)
        gmm.fit(x_all)
        bic.append(gmm.aic(x_all))
        if bic[-1] < lowest_bic:
            lowest_bic = bic[-1]
            best_gmm = gmm
            
best_gmm.fit(x_all)
gmm_train = best_gmm.predict_proba(train)
gmm_test = best_gmm.predict_proba(test)
```

```python
from sklearn import linear_model

logreg = linear_model.LogisticRegression(solver = 'saga')
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
print("Logistic Regression Accuracy: ", accuracy_score(y_test, y_pred))

print('Logistic Regression wi Std:',cross_val_score(logreg, x_std, trainLabel.values.ravel(), cv=10).mean())
print('Logistic Regression wi MMS:',cross_val_score(logreg, x_mms, trainLabel.values.ravel(), cv=10).mean())
print('Logistic Regression wi Norm:',cross_val_score(logreg, x_norm, trainLabel.values.ravel(), cv=10).mean())
print('Logistic Regression wi GMM:',cross_val_score(logreg,gmm_train, trainLabel.values.ravel(), cv=10).mean())
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
y_pred = dtree.predict(x_test)
print("Decision Tree Acuuracy: ", accuracy_score(y_test, y_pred))

print('Decision Tree wi Std:',cross_val_score(dtree, x_std, trainLabel.values.ravel(), cv=10).mean())
print('Decision Tree wi MMS:',cross_val_score(dtree, x_mms, trainLabel.values.ravel(), cv=10).mean())
print('Decision Tree wi Norm:',cross_val_score(dtree, x_norm, trainLabel.values.ravel(), cv=10).mean())
print('Decision Tree wi GMM:',cross_val_score(dtree,gmm_train, trainLabel.values.ravel(), cv=10).mean())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
print("Random Forest Acuuracy: ", accuracy_score(y_test, y_pred))

print('Random Forest wi Std:',cross_val_score(rf, x_std, trainLabel.values.ravel(), cv=10).mean())
print('Random Forest wi MMS:',cross_val_score(rf, x_mms, trainLabel.values.ravel(), cv=10).mean())
print('Random Forest wi Norm:',cross_val_score(rf, x_norm, trainLabel.values.ravel(), cv=10).mean())
print('Random Forest wi GMM:',cross_val_score(rf,gmm_train, trainLabel.values.ravel(), cv=10).mean())
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
gbc.fit(x_train, y_train)
y_pred = gbc.predict(x_test)
print("Gradient Boosting Acuuracy: ", accuracy_score(y_test, y_pred))

print('Gradient Boosting wi Std:',cross_val_score(gbc, x_std, trainLabel.values.ravel(), cv=10).mean())
print('Gradient Boosting wi MMS:',cross_val_score(gbc, x_mms, trainLabel.values.ravel(), cv=10).mean())
print('Gradient Boosting wi Norm:',cross_val_score(gbc, x_norm, trainLabel.values.ravel(), cv=10).mean())
print('Gradient Boosting wi GMM:',cross_val_score(gbc,gmm_train, trainLabel.values.ravel(), cv=10).mean())
```
