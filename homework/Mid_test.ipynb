# Pre-processing

```python
poi = train_data['poi'].astype(str).map({"True":1, "False":0})
ids = test_data['name']
train = train_data.drop(['poi'], axis=1)
df = pd.concat([train,test_data])
df = df.drop(['name'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
def na_check(df_data):
    data_na = (df_data.isnull().sum() / len(df_data)) * 100
    data_na = data_na.drop(data_na[data_na == 0].index).sort_values(ascending=False)
    missing_data = pd.DataFrame({'Missing Ratio' :data_na})
    display(missing_data.head(10))
na_check(df)
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

LEncoder = LabelEncoder()
MMEncoder = MinMaxScaler()
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = df[col].fillna(0)
    elif df[col].dtype == 'object':
#         df[col] = df.drop([col], axis=1)
        df[col] = df[col].fillna(-1)
        df[col] = LEncoder.fit_transform(list(df[col].values))
    df[col] = MMEncoder.fit_transform(df[col].values.reshape(-1, 1))
```

```python
def na_check(df_data):
    data_na = (df_data.isnull().sum() / len(df_data)) * 100
    data_na = data_na.drop(data_na[data_na == 0].index).sort_values(ascending=False)
    missing_data = pd.DataFrame({'Missing Ratio' :data_na})
    display(missing_data.head(10))
na_check(df)
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
df.corr()
```

# Model Selection

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score, classification_report

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
```

```python
train_num = poi.shape[0]
X_train = df[:train_num]
X_test = df[train_num:]
Y_train = poi.copy()
```

```python
logreg = LogisticRegression(solver='lbfgs')
logreg.fit(X_train, Y_train)
Y_pred = cross_val_predict(logreg, X_train, Y_train, cv=10)
print("Accuracy: ", accuracy_score(Y_train, Y_pred))
print (classification_report(Y_train, Y_pred))
```

```python
logreg_pred = logreg.predict_proba(X_test)[:,1]
sub = pd.DataFrame({'name': ids, 'poi': logreg_pred})
# sub['poi'] = sub['poi'].map(lambda x:1 if x>0.6 else 0) 
sub.to_csv('poi_logreg.csv', index=False) 
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)
Y_pred = cross_val_predict(dtc, X_train, Y_train, cv=10)
print("Accuracy: ", accuracy_score(Y_train, Y_pred))
print (classification_report(Y_train, Y_pred))
```

```python
dtc_pred = dtc.predict_proba(X_test)[:,1]
sub = pd.DataFrame({'name': ids, 'poi': dtc_pred})
# sub['poi'] = sub['poi'].map(lambda x:1 if x>0.6 else 0) 
sub.to_csv('poi_dtc.csv', index=False) 
```

```python
gdbt = GradientBoostingClassifier()
gdbt.fit(X_train, Y_train)
Y_pred = cross_val_predict(gdbt, X_train, Y_train, cv=10)
print("Accuracy: ", accuracy_score(Y_train, Y_pred))
print (classification_report(Y_train, Y_pred))
```

```python
gdbt_pred = gdbt.predict_proba(X_test)[:,1]
sub = pd.DataFrame({'name': ids, 'poi': gdbt_pred})
# sub['poi'] = sub['poi'].map(lambda x:1 if x>0.6 else 0) 
sub.to_csv('poi_gdbt.csv', index=False) 
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, Y_train)
Y_pred = cross_val_predict(rf, X_train, Y_train, cv=10)
print("Accuracy: ", accuracy_score(Y_train, Y_pred))
print (classification_report(Y_train, Y_pred))
```

```python
rf_pred = rf.predict_proba(X_test)[:,1]
sub = pd.DataFrame({'name': ids, 'poi': rf_pred})
# sub['poi'] = sub['poi'].map(lambda x:1 if x>0.6 else 0) 
sub.to_csv('poi_rf.csv', index=False)
```

# Stacking

```python
from mlxtend.classifier import StackingClassifier
```

```python
# meta_estimator = LogisticRegression(solver='newton-cg', C=0.001, penalty='l2')
meta_estimator = GradientBoostingClassifier(loss='deviance', learning_rate=0.15, min_samples_split=0.24545454545454548, min_samples_leaf=0.1, max_depth=3, max_features='sqrt', criterion='friedman_mse', subsample=0.85, n_estimators=10)
# meta_estimator = RandomForestClassifier(criterion='gini', max_depth=4, max_features='auto', n_estimators=200, random_state=42)
stacking = StackingClassifier(classifiers=[logreg, gdbt, rf], meta_classifier=meta_estimator
                              , use_probas=True, average_probas=False
                             )
stacking.fit(X_train, Y_train)
Y_pred = cross_val_predict(rf, X_train, Y_train, cv=10)
print("Accuracy: ", accuracy_score(Y_train, Y_pred))
print (classification_report(Y_train, Y_pred))
```

```python
stacking_pred = stacking.predict_proba(X_test)[:,1]
sub = pd.DataFrame({'name': ids, 'poi': stacking_pred})
# sub['poi'] = sub['poi'].map(lambda x:1 if x>0.6 else 0) 
sub.to_csv('poi_stacking.csv', index=False)
```
