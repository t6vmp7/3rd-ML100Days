```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame()
data["國家"] = ['American', 'Taiwan', 'China', 'Japan']
data["人口"] = ['1000', '2000', '1500', '1800']
```

```python
p_max = data[["人口"]].values.max()
data["國家"][data["人口"] == p_max]
```
