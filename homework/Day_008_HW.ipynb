```python
app_train_df = app_train.dtypes.reset_index()
app_train_df.columns = ["Column", "Column Type"]
```

```python
count = 0
for column, dtype in zip(app_train_df["Column"], app_train_df["Column Type"]):
    if dtype == "float64":
        count+=1
        if count <= 4:
            # 平均數
            print(f"{column} 平均數 : {app_train[column].mean()}")
            # 標準差
            print(f"{column} 標準差 : {app_train[column].std()}")
            # 直方圖
            data = np.log10(app_train[column]+1e-8)
            data.plot.hist(bins = 50)
            plt.title(column)
            plt.show()
```
