# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def ReLU(x):
    if x>0:
        return x
    else:
        return 0

def dReLU(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
# x = plt.linspace(-10,10,100)
# plt.plot(x, ReLU(x), 'b', label='linspace(-10,10,10)')
# plt.grid()
# plt.title('ReLU Function')
# plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=15)
# plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
# plt.gca().yaxis.set_major_locator(plt.MultipleLocator(0.1))

x =  np.arange(-10,10,0.01)
y=[]
for _ in x:
    y.append(ReLU(_))

plt.plot(x, y)
plt.grid(color='b' , linewidth='0.1' ,linestyle='--')
```

```python
Y=[]
for _ in x:
    Y.append(dReLU(_))

plt.plot(x, Y)
plt.grid(color='r' , linewidth='0.1' ,linestyle='--')

plt.show()
```
