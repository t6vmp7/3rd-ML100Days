### 2. 分類問題與回歸問題分別可用的目標函數有哪些？

Ans: 
    1) 當模型作過適當的訓練後，給予新的數據集也能適用
    2) 分類：準確率，錯誤率，查準率（precision）、查全率（recall）以及P-R曲線和ROC曲線
       回歸：均方誤差（mean squared error），絕對誤差（absolute Loss），決定係數（coefficient of determination ）以及Huber Loss
